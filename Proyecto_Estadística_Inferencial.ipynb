### $\color{turquoise}{\text{Estimador por M칠todo de Momentos}}$

El m칠todo de momentos es una **t칠cnica estad칤stica** que se utiliza para estimar p치rametros poblacionales. Consiste en **igualar** los **momentos muestrales** con los **momentos poblacionales**.  
El m칠todo fue introducido en 1887 por **Panfnuti Chebyshov**

El objetivo de este m칠todo es estimar los par치metros usando lo momentos teoricos. Sea $X$ una v.a. con par치metro $(\theta_1,\theta_2,..,\theta_p)$

$F(X|\theta), \quad\theta^T=(\theta_1,\theta_2,..,\theta_p)$. Donde $\theta$ es el vector columna que contiene los p-par치metros a ser estimados. El $k-칠simo$ momento al origen de $x$ se define como:


$$\mu'_k=\mathbb{E}(x^k|\theta)=\mathbb{E}(x^k)$$

**Nota:**
* Una desvetaja que tiene este m칠todo es que algunas distribuciones no tienen segundo momento, por lo que no es aplicable para todas las distribuciones.

Considerando una distribuci칩n exponencial con parametro $\theta$, sabemos que por definici칩n la media te칩rica es:

$$E[X] = \theta$$

Tomando en cuenta el **m칠todo de momentos**,  igualamos la media te칩rica con la media muestral, dicho de otra forma.

$$\hat{\theta} = \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i$$

$$\bar{X} = \frac{7270}{20}$$

**$$\bar{X} = 363.5 $$**

Por lo tanto, nuestro estimador **$\theta$** calculado por el m칠todo de momentos es:

**$$\hat{\theta} = 363.5$$**

### $\color{turquoise}{\text{Estimador de M치xima Verosimilitud}}$

Este procedimental t칠cnico se emplea para todo tipo de datos: completos, agrupados y los modificados (truncados, censurados, y la comnbinaci칩n de ambos), as칤 mismo, es aplicable tanto para variables discretas y continuas.

Para definir **EMV** se supone un conjunto de datos que consiste en n-eventos $A_1, A_2,...,A_n$. Donde $A_j$ es cualquier conjunto de datos del j-칠simo evento.

Entonces del evento $A_j$ se puede observar v.a.i.i.d. $X_j$, por lo tanto, el EMV se usar치 para estimar el **par치metro que m치ximiza** la probabilidad de obtener los datos observados

***Pasos***

$\color{gray}{\text{1.- Se establece la funci칩n de verosimilitud}}$  

**Funci칩n de Verosimilitud**  
Sea $X_1, X_2, \dots, X_n$ una muestra aleatoria con $f(x;\theta)$. La funci칩n de verosimilitud (likelihood) es:

$$f(X_1, X_2, \dots, X_n; \theta)=f(X_1,;\theta)\cdots(x_n;\theta)$$
Entonces:
$$L(x_1, \dots, x_n; \theta)=\prod_{i=1}^nf(x_i,\theta)$$

Donde $f(X_1, X_2, \dots, X_n; \theta)$ es la funci칩n de densidad o de probabilidad dependiendo del tipo de variable

Una vez definida la funcio칩n de verosimilitud el procedimiento es muy senvillo para obtener el **EMV**

 
$\color{gray}{\text{2.- Se establece la funci칩n log-verosimilitud, esta se obtiene aplicando logaritmo al funci칩n de verosimilitud}}$ 

$$ \ln L(x1,\dots, x_n; \theta) $$  

$\color{gray}{\text{3.- Se obtiene la derivada con respecto a}\,\,\theta \,\, \text{e igualamos a 0, posteriormente despejamos}\,\,\theta}$

$$ \frac{d L(x1,\dots, x_n; \theta)=}{d \theta} =0 $$  

$\color{gray}{\text{4.- Finalmente se tendra que comprobar si es un m치ximo a traves de la segunda derivada}}$ 
$$ \frac{d^2 L(x1,\dots, x_n; \theta)}{d^2 \theta} =0 $$  
$$\frac{d^2}{d^2\theta}l(\theta)|_{\theta=\hat\theta}$$

### $\color{turquoise}{\text{Propiedades de los Estimadores}}$

$\color{gray}{\text{Insesagado}}$  

Dicho de otra forma un estimador **insesgado** es aquel que tiene un valor esperado igual al par치metro que se quiere estimar.

$\color{gray}{\text{Consistentes}}$  

Un estimador es consistente cuando se aproxima cada vez m치s al valor del par치metro a medida que aumenta el tama침o de la muestra.  
A mayor tama침o de la muestra, menor es la probalidad de que el estimador este muy lejos del valor real, la precisi칩n aumenta a medida que el tama침o de la muestra tambi칠n lo hace.

$\color{gray}{\text{Invarianza}}$  

Un estimador es invariante cuando el estimador de la funci칩n del par치metro coincide con la funci칩n del estimador del par치metro. 

$\color{gray}{\text{Suficiente}}$  

Un estimador es suficiente cuando resume toda la informaci칩n relevante de una muestra, sin que ning칰n otro estimador pueda aportar m치s informaci칩n. 

$\color{gray}{\text{Eficiente}}$  

En estad칤stica, un estimador es eficiente cuando tiene la menor varianza posible, lo que indica que es m치s preciso. Un estimador eficiente es el "mejor posible" o "칩ptimo" de un par치metro de inter칠s. 

# $$\color{forestgreen}{\text{INTERVALOS DE CONFIANZA (IC)}}$$

### $\color{limegreen}{\text{IC para la media de una distribuci칩n nomal }(\sigma\,\,conocida)}$

Sea $X_1, \dots, X_n$ una m.a. $X_i \sim \mathcal{N}(\mu, \sigma^2)$. Entonces:

- $\bar{X} \sim \mathcal{N}(\mu, \frac{\sigma^2}{n})$
- $Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \sim \mathcal{N}(0,1)$

***Nivel de significancia (error): $\alpha$***

El intervalo de confianza se basa en que:
$$
\mathbb{P}\left( -z_{\alpha/2} < \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} < z_{\alpha/2} \right)
= \mathbb{P}\left( \bar{X} - z_{\alpha/2} \frac{\sigma}{\sqrt{n}} < \mu < \bar{X} + z_{\alpha/2} \frac{\sigma}{\sqrt{n}} \right)
=1-\alpha
$$  
  


***El intervalo del $(1 - \alpha) \cdot 100\%$ de confianza para $\mu$ (con $\sigma$ conocida) es:***

$$
\left( \bar{X} - z_{\alpha/2} \frac{\sigma}{\sqrt{n}}, \quad \bar{X} + z_{\alpha/2} \frac{\sigma}{\sqrt{n}} \right)
$$
donde

- $\sigma$ = desviaci칩n est치ndar  
- $\frac{\sigma}{\sqrt{n}}$ = error est치ndar de la media  
- $z_{\alpha/2}$ determina el nivel de confianza  
- El intervalo est치 centrado en $\bar{X}$

*Observaci칩n*

- A mayor $n$, menor es el error est치ndar, y el intervalo de confianza es m치s peque침o.
- Esto hace que se acerque m치s a la media $\mu$, es decir, que haya menor맜ariaci칩n.    

$\color{darkgray}{\text{Ejemplo:}}$  
Para tratar de estimar la media de consumo por cliente en un gran restaurante, se reuniero datos de una muestra de 49 clientes durante 3 semanas  

**a)** Supon que la **desviaci칩n estandar de la poblaci칩n** es de $\ 2.5$ 쮺u치l es el error estandar de la media?  
**b)** Con un nivel de confianza del $\ 95%$ 쮺u치l es el margen de error?  
$1-\alpha=0.95$, implica que $z_{\alpha/2}=z_.025=1.96$  

**c)** Si la **media de la muestra** es de $\ 22.60$ 쮺u치l es el intervalo de confianza al 95% para la media de la poblaci칩n?  

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import scipy.stats as chi2
import scipy.stats as norm

#Datos
n=49
sigma=2.5
media_muestral=22.6 #X barra
confianza=.95

#a) Error estandar de la media 
error_estandar=sigma/np.sqrt(n)
error_estandar
print(f" a) Error estandar de la media: {error_estandar:.2f}")

#b) Margen cr칤tico (Z de .025)
z=st.norm.ppf(1-(1-confianza)/2)
margen_error=z*error_estandar
margen_error
print(f" b) Margen de error de la media: {margen_error:.2f}")

#c) Limite inferior
limite_inferior= media_muestral-margen_error
limite_superior= media_muestral+margen_error
print(f" c) Intervalo de confianza al 95%: ({limite_inferior:.2f},{limite_superior:.2f})")

```

### $\color{limegreen}{\text{IC para la media de una distribuci칩n nomal }(\sigma\,\,desconocida)}$

Sea $X_1, \dots, X_n$ una m.a. $X_i \sim \mathcal{N}(\mu, \sigma^2)$. Entonces:

- $\bar{X} \sim \mathcal{N}(\mu, \frac{\sigma^2}{n})$
- $Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \sim \mathcal{N}(0,1)$

Si el tama침o de la muestra es menor a 30, se utiliza la distribucion t:

$\color{darkgray}{\text{Estadistico }t}$  

Definimos el estadistico t de la siguiente manera:
$$ T=\frac{\bar{x}-\mu}{s/\sqrt{n}} \sim {t_(n-1)} $$

$\color{darkgray}{\text{El intervalo }(1-\alpha)\cdot100\%\text{ de confianza para }\mu\text{ con } \sigma\,\,desconocida}$ 

$$
\left( \bar{X} - t_{\alpha/2} \frac{s}{\sqrt{n}}, \quad \bar{X} + t_{\alpha/2} \frac{s}{\sqrt{n}} \right)
$$
donde

- $\sigma$ = desviaci칩n est치ndar muestral
- $\frac{s}{\sqrt{n}}$ = error est치ndar de la media  
- $t_{\alpha/2}$ determina el nivel de confianza  
- El intervalo est치 centrado en $\bar{X}$

*Observaci칩n*

- A mayor $n$, menor es el error est치ndar, y el intervalo de confianza es m치s peque침o.
- Esto hace que se acerque m치s a la media $\mu$, es decir, que haya menor맜ariaci칩n.

$\color{darkgray}{\text{Ejemplo}}$ 
Supongamos que tenemos los siguientes datos: 

$$ datos=[45,55,67,45,68,79,98,87,84,82] $$

* Calcular un intervalo de confianza para la media.


```python
#Datos del ejemplo

data=[45,55,67,45,68,79,98,87,84,82]
confidence=0.95
gl=len(data)-1 #Grados de libertad

#Media y error estandar
mean=np.mean(data)
error_est=st.sem(data)


#Intervalo de confianza usando t de Student
intervalo=st.t.interval(confidence, gl, loc=mean, scale=error_est)
print(f"Media muestral: {mean:.2f}")
print(f"Error estandar: {error_est:.2f}")
print(f"Intervalo de confianza al 95% es: {intervalo}")
```

# $$\color{orange}{\text{Distribuciones Muestrales}}$$

### $\color{limegreen}{\text{Distribuci칩n }t\,\,de\,\,Student}$

Sean:
- $X \sim \mathcal{N}(0,1)$  
- $Y \sim \chi^2(n - 1)$  

Entonces:

$$
T = \frac{X \cdot \sqrt{n}}{\sqrt{\dfrac{Y}{n - 1}}} \sim t(n - 1)
$$

Gr치ficamente, la distribuci칩n *t* tiene forma de campana, similar a la normal est치ndar, pero con colas m치s gruesas. A medida que $n \to \infty$, se aproxima a una normal est치ndar:

$$
\lim_{n \to \infty} t(n) = \mathcal{N}(0,1)
$$

---

> 游늷 Esta distribuci칩n es 칰til cuando se desconoce la varianza poblacional y se utiliza una estimaci칩n muestral.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Valores de x para graficar
x = np.linspace(-5, 5, 500)

# Distribuciones t con diferentes grados de libertad
dfs = [1, 3, 10, 30]
colors = ['red', 'orange', 'green', 'blue']

plt.figure(figsize=(10, 6))

# Graficar curvas t
for df, color in zip(dfs, colors):
    plt.plot(x, t.pdf(x, df), label=f't(df={df})', color=color)

# Graficar la normal est치ndar
plt.plot(x, norm.pdf(x), 'k--', label='Normal est치ndar')

# Formato de la gr치fica
plt.title('Comparaci칩n entre la distribuci칩n t y la normal est치ndar', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

### $\color{limegreen}{\text{Distribuci칩n }F\,\,de\,\,Fisher}$

Sea:

$$
F = \frac{\chi^2_{(m-1)} / (m - 1)}{\chi^2_{(n-1)} / (n - 1)} \sim F(m - 1,\ n - 1)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f, chi2

# Rango de valores para las gr치ficas
x_f = np.linspace(0, 5, 500)
x_chi = np.linspace(0, 20, 500)

# --- Distribuci칩n F de Fisher ---
df1, df2 = 5, 10
f_dist = f.pdf(x_f, df1, df2)

plt.figure(figsize=(12, 5))

# Gr치fica de F
plt.subplot(1, 2, 1)
plt.plot(x_f, f_dist, label=f'F({df1}, {df2})', color='purple')
plt.title('Distribuci칩n F de Fisher', fontsize=14)
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
```

### $\color{limegreen}{\text{Distribuci칩n }Ji-cuadrada\,\,(Chi-cuadrada)}$

Si:

- $X \sim \mathcal{N}(0,1)$  
- Entonces: $X^2 \sim \chi^2(1)$

Y si:

$$
\sum_{i=1}^{n} X_i^2 \sim \chi^2(n)
$$

> 游늷 La distribuci칩n $\chi^2(k)$ surge como la suma de los cuadrados de $k$ variables normales est치ndar independientes.


```python
# --- Distribuciones Ji-cuadrada ---
df_chi_list = [1, 3, 6, 10]

plt.subplot(1, 2, 2)
for df in df_chi_list:
    plt.plot(x_chi, chi2.pdf(x_chi, df), label=f'픦({df})')
plt.title('Distribuciones Ji-cuadrada', fontsize=14)
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
